#### Normalized Difference Vegetation Index (NDVI)

https://gisgeography.com/ndvi-normalized-difference-vegetation-index/

Normalized Difference Vegetation Index (NDVI) quantifies vegetation by measuring the difference between near-infrared (which vegetation strongly reflects) and red light (which vegetation absorbs). NDVI always ranges from -1 to +1. 

Overall, NDVI is a standardized way to measure healthy vegetation. When you have high NDVI values, you have healthier vegetation. When you have low NDVI, you have less or no vegetation. 

For example, when you have negative values, it’s highly likely that it’s water. On the other hand, if you have a NDVI value close to +1, there’s a high possibility that it’s dense green leaves.

But when NDVI is close to zero, there isn’t green leaves and it could even be an urbanized area.


```python
# NDVI
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))

# San Juan
ax1.set_title('San Juan::NDVI per week')
ax1.grid()

ax1.set_xlabel('year')
ax1.set_ylabel('NDVI')

ax1.plot(df_train_sj['week_start_date'], df_train_sj['ndvi_ne'], label='NE')
ax1.plot(df_train_sj['week_start_date'], df_train_sj['ndvi_nw'], label='NW')
ax1.plot(df_train_sj['week_start_date'], df_train_sj['ndvi_se'], label='SE')
ax1.plot(df_train_sj['week_start_date'], df_train_sj['ndvi_sw'], label='SW')

ax1.legend()

# Iquitos
ax2.set_title('Iquitos::NDVI per week')
ax2.grid()

ax2.set_xlabel('year')
ax2.set_ylabel('NDVI')

ax2.plot(df_train_iq['week_start_date'], df_train_iq['ndvi_ne'], label='NE')
ax2.plot(df_train_iq['week_start_date'], df_train_iq['ndvi_nw'], label='NW')
ax2.plot(df_train_iq['week_start_date'], df_train_iq['ndvi_se'], label='SE')
ax2.plot(df_train_iq['week_start_date'], df_train_iq['ndvi_sw'], label='SW')

ax2.legend()

plt.show()
```

### Cleaning



```python
# function to fill values with the mean
def fillna_with_mean(df):
    for col in df:
        if col not in ['city','week_start_date']:
            df[col].fillna(df[col].mean(), inplace=True)
        
fillna_with_mean(df_train_sj)
fillna_with_mean(df_train_iq)
```
